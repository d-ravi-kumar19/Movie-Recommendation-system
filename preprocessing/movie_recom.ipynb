```python
movie_credits = pd.read_csv(tmdb_path + 'tmdb_5000_credits.csv')
movie_credits.head()
```

```python
def extract_cast(obj):
    count = 0
    l =[]
    for i in literal_eval(obj):
        if count <3:
            l.append(i['name'])
            count +=1
        else:
            break
    return l

def extract_director(obj):
    l =[]
    for i in literal_eval(obj):
        if i['job'] =='Director':
            l.append(i['name'])
            break
    return l

movie_credits['cast'] = movie_credits['cast'].apply(extract_cast)
movie_credits['crew'] = movie_credits['crew'].apply(extract_director)

movie_credits['cast'] = movie_credits['cast'].apply(lambda x : [item.replace(' ', '') for item in x])
movie_credits['crew'] = movie_credits['crew'].apply(lambda x : [item.replace(' ', '') for item in x])
movie_credits['crew'] = movie_credits['crew'].apply(lambda x : x[0] if x else '')
movie_credits.rename(columns={'crew':'director'},inplace=True)
```
