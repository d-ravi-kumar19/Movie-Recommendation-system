```python
import requests

# Define the base URL
base_url = "http://202.63.117.72/result_dec_2023/31/index.php"

# Define the range of hall ticket numbers
start_number = 100521729001
end_number = 100521729002

# Loop through the hall ticket numbers
for hall_ticket_number in range(start_number, end_number + 1):
    # Prepare the data for the POST request
    data = {'htno': str(hall_ticket_number)}

    # Make a POST request to the server
    response = requests.post(base_url, data=data)

    # Print the content of the response
    print(f"Hall Ticket Number: {hall_ticket_number}")
    print(response.text)
    print("--------------------")
```

```python
import requests
from bs4 import BeautifulSoup

def fetch_result(htno):
    # Enter Hall Ticket Number
    payload = {'htno': htno}
    url = 'http://202.63.117.72/result_dec_2023/31/index.php'
    response = requests.post(url, data=payload)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        result_page_html = response.text

        # Parse the result page with BeautifulSoup
        soup = BeautifulSoup(result_page_html, 'html.parser')
        # print(soup)
        # Extract the necessary information
        name_element = soup.find('tr', bgcolor="#5278a7")
        if name_element:
            name = name_element.find('td', height="27").get_text(strip=True)
        else:
            name = "Name not found"

        cgpa_element = soup.find('td', class_='style19')
        cgpa = cgpa_element.get_text(strip=True) if cgpa_element else "CGPA not found"

        # Print the results
        print(f"Hall Ticket Number: {htno}")
        print(f"Name: {name}")
        print(f"CGPA: {cgpa}")

        # Extract and print the subject details
        subject_table = soup.find('table', width="662", border="1", align="center")
        if subject_table:
            subjects = subject_table.find_all('tr')[1:]  # Skip the header row
            for subject in subjects:
                subject_name = subject.find_all('td')[1].get_text(strip=True)
                grade = subject.find_all('td')[3].get_text(strip=True)
                print(f"Subject: {subject_name}, Grade: {grade}")
    else:
        print(f"Error: Unable to fetch data for Hall Ticket Number {htno}")

# Example: Replace '100521729001' with the desired hall ticket number
fetch_result('100521729003')
```
